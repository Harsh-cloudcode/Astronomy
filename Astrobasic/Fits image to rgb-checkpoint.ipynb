Horse head image


```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

# Load the FITS image
fits_image = fits.open('HorseHead.fits')
data = fits_image[0].data  # Assuming your image is in the primary HDU

# Normalize the data (optional but recommended)
data_min = np.min(data)
data_max = np.max(data)
data_normalized = (data - data_min) / (data_max - data_min)

# Define color mapping (e.g., using a colormap)
# You can choose any colormap you like, or create your custom colormap.
# Here's an example using a colormap called 'viridis':
cmap = plt.cm.viridis

# Create the RGB image by mapping the data to the color channels
# You can adjust the colormap and scaling as needed to highlight features.
rgb_image = cmap(data_normalized)

# Display the RGB image
plt.imshow(rgb_image)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```
